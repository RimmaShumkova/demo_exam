## Обработка вещественных признаков (заполнение пропусков) <a id='toc1_4_'></a>

Воспользуемся статистическими методами для заполнения пропущенных значений и найдем лучший метод для заполнения по KL-дивергенции.

```python
def fill_missing_kl(df, column):
    candidates = {
        'mean': df[column].mean(),
        'median': df[column].median(),
        'mode': df[column].mode()[0]
    }
    
    kl_values = {}
    
    hist_orig, bins = np.histogram(df[column].dropna(), bins='auto', density=True) 

    for name, val in candidates.items():
        df_copy = df.copy()
        df_copy[column] = df_copy[column].fillna(val)
        hist_new, _ = np.histogram(df_copy[column], bins=bins, density=True)
        hist_new += 1e-10
        hist_orig_safe = hist_orig + 1e-10
        kl = entropy(hist_orig_safe, hist_new)
        kl_values[name] = kl

        plt.figure()
        plt.hist(df[column].dropna(), bins=bins, alpha=0.5, label='Original')
        plt.hist(df_copy[column], bins=bins, alpha=0.5, label=f'Fill={name} ({val:.2f})')
        plt.legend()
        plt.title(f'{name}: KL={kl:.4f}')
        plt.show()

    best_method = min(kl_values, key=kl_values.get)
    best_val = candidates[best_method]
    df[column] = df[column].fillna(best_val)

    print(f'Выбрано для заполнения: {best_method} (значение={best_val:.4f})')
    return df

X_train_stat = X_train.copy()
X_train_stat = fill_missing_kl(X_train_stat, 'Скорость перехода через портал')
```

Применяем полученный результат к тестовым данным.

### Оценка качества заполнения пропусков машинным методом

**NRMSE** — это нормализованная среднеквадратичная ошибка. Она показывает, насколько сильно восстановленные (заполненные) значения отличаются от истинных.

```python
def nrmse(true, imputed):
    rmse = np.sqrt(mean_squared_error(true, imputed))
    return rmse / (true.max() - true.min()) # нормировка через размах max-min истинных значений
```

Убираем все пропуски из обучающей выборки и заполняем ее "искусственными" пропусками.

```python
clean_data = X_train[num_cols].dropna().copy()

mask = np.random.rand(*clean_data.shape) < 0.1
data_missing = clean_data.mask(mask) # в повреждениях ставим NaN

data_missing.isna().sum()
```

Применяем KNNImputer для искусственных пропусков.

```python
imputer_test = KNNImputer(n_neighbors=5)
data_filled = imputer_test.fit_transform(data_missing)
```

Считаем средний NRMSE по каждому признаку и выводим получившийся результат.

```python
nrmse_scores = {}
for i, col in enumerate(clean_data.columns):
    missing_mask = mask[:, i] # для каждого столбца берётся булев вектор (какие строки испорчены в столбце)
    if missing_mask.sum() > 0: # сравниваем элементы, которые были закрыты при порче
        score = nrmse(clean_data[col].values[missing_mask], 
                      data_filled[missing_mask, i])
        nrmse_scores[col] = score
```

```python
nrmse_df = pd.DataFrame.from_dict(nrmse_scores, orient='index', columns=['NRMSE']).sort_values('NRMSE')
display(nrmse_df)

print(f"Средний NRMSE по всем признакам: {nrmse_df['NRMSE'].mean():.4f}")
```

Почти все признаки имеют NRMSE < 0.01, что означает, что KNNImputer восстановил значения очень точно.

**Вывод:** средний NRMSE = 0.0189 — очень хороший результат. Это говорит о том, что KNNImputer восстанавливает пропуски почти идеально для большинства признаков.

## Детекция выбросов и аномалий в данных <a id='toc1_6_'></a>

### Поиск выбросов в данных с помощью **IQR** и **Robust Z-score.**

Коэффициент **k=1.5** для IQR - это стандартный коэффициент в статистике для умеренных выбросов.

В Robust Z-score **threshold=3** соответствует правилу трех сигм в статистике. Robust Z-score использует MAD вместо std, что делает его устойчивее к выбросам.

```python
X = X_train_knn.copy()

def iqr_outliers(data, k=1.5):
    q1 = data.quantile(0.25)
    q3 = data.quantile(0.75)
    iqr = q3 - q1
    lower = q1 - k * iqr
    upper = q3 + k * iqr
    mask = (data < lower) | (data > upper)
    return mask

def robust_z_score_outliers(data, threshold=3):
    med = data.median()
    mad = np.median(np.abs(data - med)) # медианное абсолютное отклонение
    if mad == 0:
        mad = np.std(data)
    rzs = 0.6745 * (data - med) / mad
    return rzs.abs() > threshold
```

Для каждого столбца построим диаграмму "Ящик с усами" и по ней определим, есть ли выбросы в данном признаке.

```python
outlier_rows_per_column = {}

plt.figure(figsize=(12, max(4, len(X.columns)*1.2)))
for col in X.columns:
    plt.figure(figsize=(6,2))
    sns.boxplot(x=X[col])
    plt.title(f"Boxplot: {col}")
    plt.tight_layout()
    plt.show()
    
    mask_iqr = iqr_outliers(X[col], k=1.5)
    mask_rz = robust_z_score_outliers(X[col], threshold=3)
    intersect_idx = set(X.index[mask_iqr & mask_rz])
    outlier_rows_per_column[col] = intersect_idx
    print(f"{col}: IQR выбросов = {mask_iqr.sum()}, RobustZ выбросов = {mask_rz.sum()}, пересечение = {len(intersect_idx)}")
```

Находим строки для удаления выбросов на основе пересечения.

```python
rows_to_drop_iqr_rz = set().union(*outlier_rows_per_column.values())
print(f"Количество строк для удаления по IQR & RobustZ: {len(rows_to_drop_iqr_rz)}")
```

Удаляем выбросы на основе пересечения строк, в которых оба метода обнаружили выбросы.

Для подбора оптимального epsilon для DBSCAN создадим функцию, которая подберет этот параметр.

```python
def auto_dbscan(X, min_samples=5, plot=True, quantile=0.95):

    # вычисляем расстояние до k-го соседа
    neigh = NearestNeighbors(n_neighbors=min_samples)
    neigh.fit(X)
    distances, _ = neigh.kneighbors(X)
    k_distances = np.sort(distances[:, -1])
    eps = np.quantile(k_distances, quantile) # берем квантиль как значение eps

    model = DBSCAN(eps=eps, min_samples=min_samples)
    model.fit(X)
    labels = (model.labels_ == -1).astype(int)

    print(f"Выбран eps={eps:.3f}, найдено {labels.sum()} аномалий")

    return model, labels, eps
    
clf_dbscan, labels_dbscan, eps_used = auto_dbscan(X_scaled, min_samples=5, plot=True, quantile=0.95)
```

```python
labels_iforest = pd.Series(clf_iforest.labels_, index=X_scaled.index)
labels_ocsvm = pd.Series(clf_ocsvm.labels_, index=X_scaled.index)
labels_dbscan = pd.Series((clf_dbscan.labels_ == -1).astype(int), index=X_scaled.index)

print("Количество аномалий:")
print(f"IsolationForest: {labels_iforest.sum()}")
print(f"OneClassSVM:    {labels_ocsvm.sum()}")
print(f"DBSCAN:         {labels_dbscan.sum()}")
```

### Визуализация аномалий на двумерной плоскости

```python
pca = PCA(n_components=2, random_state=42)
X_pca = pd.DataFrame(pca.fit_transform(X_scaled), index=X_scaled.index, columns=['PC1','PC2'])

def plot_anomalies_2d(pca_df, labels, title):
    plt.figure(figsize=(6,5))
    plt.scatter(pca_df[labels==0]['PC1'], pca_df[labels==0]['PC2'], s=20, alpha=0.6, label='Normal')
    plt.scatter(pca_df[labels==1]['PC1'], pca_df[labels==1]['PC2'], s=30, edgecolor='k', facecolors='none', label='Anomaly')
    plt.xlabel('PC1'); plt.ylabel('PC2')
    plt.title(title)
    plt.legend()
    plt.tight_layout()
    plt.show()

plot_anomalies_2d(X_pca, labels_iforest, 'IsolationForest')
plot_anomalies_2d(X_pca, labels_ocsvm, 'One-class SVM')
plot_anomalies_2d(X_pca, labels_dbscan, 'DBSCAN')
```

# Выбор моделей ML и метрик <a id='toc3_'></a>

### Выбор моделей ML:

1. **LinearRegression** - базовая линейная модель для сравнения

2. **Ridge** - регуляризованная линейная модель

3. **RandomForest** - ансамблевая модель на основе бэггинга

4. **GradientBoosting** - классический градиентный бустинг от scikit-learn

5. **CatBoost** - современный градиентный бустинг с продвинутыми алгоритмами

### Критерии выбора моделей:

- Разнообразие подходов:

1. Линейные модели: для выявления линейных зависимостей

2. Деревья решений: для улавливания нелинейностей

3. Ансамблевые методы: для повышения точности и устойчивости

- Практическая обоснованность:

1. CatBoost часто показывает state-of-the-art результаты на табличных данных

2. RandomForest устойчив к переобучению на небольших выборках

3. Линейные модели обеспечивают интерпретируемость

### Описание выбранных моделей:

- **LinearRegression** - статистический метод, моделирующий линейную зависимость между независимыми переменными (признаками) и зависимой переменной (целевым значением).

Особенности:
1. Быстрое обучение
2. Не имеет гиперпараметров для подбора
3. Интерпретируемые коэффициенты
4. Чувствительность к мультиколлинеарности и выбросам

- **Ridge** - это регуляризованная версия линейной регрессии с L2-регуляризацией, которая добавляет штраф за большие значения весов.

Особенности:
1. Эффективна при мультиколлинеарности
2. Гиперпараметр α контролирует компромисс смещение-дисперсия
3. Сохраняет все признаки (в отличие от Lasso)
4. Более устойчива к переобучению чем LinearRegression

- **RandomForest** - ансамблевый метод, использующий бэггинг (bootstrap aggregating) и случайные подпространства признаков для построения множества деревьев решений.

Как работает:
1. Бэггинг: создает множество bootstrap-выборок из обучающих данных
2. Для каждой выборки строит дерево решений с ограничениями: в каждом узле рассматривается случайное подмножество признаков (√p или log₂p); ограничение максимальной глубины дерева
3. Регрессия в деревьях: в листьях хранятся средние значения целевой переменной
4. Агрегация предсказаний: итоговое предсказание — среднее предсказаний всех деревьев

Особенности:
1. Устойчивость к переобучению (благодаря бэггингу)
2. Обработка нелинейных зависимостей
3. Оценка важности признаков
4. Мало чувствителен к выбросам
5. Параллелизуемость обучения

- **Gradient Boosting** - ансамблевый метод, который строит последовательность слабых предсказателей (обычно неглубокие деревья), где каждое следующее дерево учится на ошибках предыдущих.

Особенности:
1. Высокая предсказательная способность
2. Чувствительность к гиперпараметрам
3. Склонность к переобучению при неправильной настройке
4. Последовательное (не параллельное) обучение

- **CatBoost (Categorical Boosting)** - современная реализация градиентного бустинга, специально разработанная для работы с категориальными признаками, но эффективная и с числовыми.

Дополнительные преимущества:
1. Встроенная регуляризация (L2, bagging temperature)
2. Автоматическая обработка пропущенных значений
3. Эффективная GPU-реализация
4. Меньше требует предобработки данных

Особенности:
1. State-of-the-art на многих задачах табличных данных
2. Меньше склонен к переобучению чем другие бустинги
3. Имеет больше гиперпараметров для тонкой настройки
4. Поддерживает различные функции потерь

### Выбор метрик ML:
1. **MAE (Средняя абсолютная ошибка)** - для интерпретируемой оценки абсолютных ошибок
2. **RMSE (Среднеквадратичная ошибка)** - для чувствительной оценки больших ошибок
3. **R² (Коэффициент детерминации)** - для оценки доли объясненной дисперсии
4. **MAPE (Средняя абсолютная процентная ошибка)** - для относительной оценки ошибок

### Описание выбранных метрик:

- **MAE** - Средняя абсолютная ошибка
### MAE = (1/n) * Σ|y_i - ŷ_i|

Пояснение:
1. Сумма абсолютных значений ошибок, деленная на количество наблюдений
2. Измеряется в тех же единицах, что и целевая переменная
3. Менее чувствительна к выбросам, чем RMSE

Средняя абсолютная ошибка MAE вычисляет среднее абсолютное значение разниц между предсказанными и фактическими значениями. Она измеряется в тех же единицах, что и целевая переменная, что делает её интуитивно понятной для интерпретации. Главное преимущество MAE в её устойчивости к выбросам, поскольку она не возводит ошибки в квадрат, однако это же свойство делает её менее чувствительной к крупным ошибкам, которые могут быть критичными в некоторых приложениях. Эта метрика хорошо подходит для случаев, когда все ошибки одинаково важны.

- **RMSE** - Среднеквадратичная ошибка
### RMSE = √( (1/n) * Σ(y_i - ŷ_i)² ) = √(MSE)

Пояснение:
1. Квадратный корень из средней квадратичной ошибки
2. Измеряется в тех же единицах, что и целевая переменная
3. Более чувствительна к большим ошибкам (штрафует квадратично)

Среднеквадратичная ошибка RMSE представляет собой корень квадратный из средней квадратичной ошибки и, как и MAE, измеряется в единицах целевой переменной. Ключевое отличие - квадратичная природа этой метрики делает её более чувствительной к большим ошибкам, поскольку каждая ошибка возводится в квадрат перед усреднением. Это свойство особенно ценно в задачах, где крупные отклонения недопустимы или особенно дорого обходятся. RMSE часто используется как целевая функция при оптимизации моделей, но её интерпретация может быть менее интуитивной, чем у MAE, из-за влияния квадратичного преобразования.

- **R²** - Коэффициент детерминации
### R² = 1 - (Σ(y_i - ŷ_i)²) / (Σ(y_i - ȳ)²)
где:

Σ(y_i - ŷ_i)² - сумма квадратов остатков

Σ(y_i - ȳ)² - общая сумма квадратов

y_i - истинное значение

ŷ_i - предсказанное значение

ȳ - среднее значение целевой переменной

i = 1, ..., n - количество наблюдений

Коэффициент детерминации R² показывает, какую долю дисперсии целевой переменной объясняет модель. Значения изменяются от -∞ до 1, где 1 означает идеальное соответствие модели данным, 0 - что модель предсказывает не лучше, чем простое среднее значение, а отрицательные значения указывают на то, что модель работает хуже этого базового уровня. Эта метрика особенно полезна для сравнения разных моделей на одной и той же задаче, поскольку она нормирована и позволяет оценить общую объясняющую способность модели независимо от масштаба данных. Однако R² не показывает абсолютную величину ошибок и может быть обманчиво высоким при наличии выбросов.

- **MAPE** - Средняя абсолютная процентная ошибка
### MAPE = (1/n) * Σ|(y_i - ŷ_i) / y_i| * 100%

Пояснение:
1. Выражается в процентах
2. Позволяет сравнивать модели на данных разного масштаба
3. Не рекомендуется при значениях целевой переменной близких к нулю

Средняя абсолютная процентная ошибка MAPE выражает среднюю абсолютную ошибку в процентном отношении к фактическим значениям. Это делает её безразмерной метрикой, которая позволяет сравнивать качество моделей на задачах с разными масштабами данных - например, сравнивать точность прогноза цен на недвижимость и цен на акции. MAPE особенно популярна в бизнес-аналитике и экономическом прогнозировании, где ошибки традиционно выражают в процентах. Однако у этой метрики есть существенный недостаток: она не определена при нулевых значениях целевой переменной и может давать искажённые результаты при значениях, близких к нулю, а также асимметрично наказывает за заниженные и завышенные прогнозы.

Определим сетку гиперпараметров

```python
def get_model_params(trial, model_name):
    if model_name == 'Ridge':
        return {
            'alpha': trial.suggest_float('alpha', 0.01, 100.0, log=True),
            'solver': trial.suggest_categorical('solver', ['auto', 'svd', 'cholesky', 'lsqr'])
        }
    elif model_name == 'RandomForest':
        return {
            'n_estimators': trial.suggest_int('n_estimators', 50, 300),
            'max_depth': trial.suggest_int('max_depth', 3, 15),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 10),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 5),
            'max_features': trial.suggest_categorical('max_features', ['sqrt', 'log2', None])
        }
    elif model_name == 'GradientBoosting':
        return {
            'n_estimators': trial.suggest_int('n_estimators', 50, 300),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
            'max_depth': trial.suggest_int('max_depth', 3, 10),
            'subsample': trial.suggest_float('subsample', 0.5, 1.0),
            'min_samples_split': trial.suggest_int('min_samples_split', 2, 20),
            'min_samples_leaf': trial.suggest_int('min_samples_leaf', 1, 10)
        }
    elif model_name == 'CatBoost':
        return {
            'iterations': trial.suggest_int('iterations', 100, 500),
            'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
            'depth': trial.suggest_int('depth', 3, 10),
            'l2_leaf_reg': trial.suggest_float('l2_leaf_reg', 1, 10),
            'border_count': trial.suggest_int('border_count', 32, 255),
            'bagging_temperature': trial.suggest_float('bagging_temperature', 0, 1)
        }
    else:  
        return {}
```

# Обучение моделей ML и подбор гиперпараметров <a id='toc4_'></a>

```python
from sklearn.base import clone
```

```python
best_models = {}
best_scores = {}
pca_versions = {
    'std': X_train_pca_std,
    'minmax': X_train_pca_minmax, 
    'robust': X_train_pca_robust
}

for model_name in models.keys():
    model_best_score = float('inf')
    model_best_version = None
    model_best_params = None
    model_best_instance = None
    
    print(f"\nПодбор гиперпараметров для {model_name}")
    
    for version_name, X_train_data in pca_versions.items():
        print(f"\nВерсия данных: {version_name}")
        
        def objective(trial):
            params = get_model_params(trial, model_name)
            
            model = clone(models[model_name]) # новый экземпляр модели на основе базового
            model.set_params(**params)
            
            # кросс-валидация
            kf = KFold(n_splits=2, shuffle=True, random_state=42) # исправлено на 2
            scores = cross_val_score(model, X_train_data, y_train, 
                                     cv=kf, scoring='neg_root_mean_squared_error')
            return scores.mean()
        
        study = optuna.create_study(direction='maximize')
        study.optimize(objective, n_trials=3, show_progress_bar=True) # исправлено на 3
        
        current_rmse = -study.best_value
        
        print(f"Лучшие параметры: {study.best_params}")
        print(f"Лучшее RMSE: {current_rmse:.4f}")
        
        if current_rmse < model_best_score:
            model_best_score = current_rmse
            model_best_version = version_name
            model_best_params = study.best_params
            
            final_model = clone(models[model_name])  # новый экземпляр для финального обучения
            final_model.set_params(**study.best_params)
            final_model.fit(X_train_data, y_train)
            model_best_instance = final_model
    
    print(f"\nЛучшая версия для {model_name}: {model_best_version} с RMSE: {model_best_score:.4f}")
    
    best_models[model_name] = {
        'model': model_best_instance,
        'version': model_best_version,
        'params': model_best_params
    }
    best_scores[model_name] = model_best_score
```

# Вычисление метрик на новых данных <a id='toc5_'></a>

```python
pca_data_mapping = {
    'std': (X_train_pca_std, X_test_pca_std),
    'minmax': (X_train_pca_minmax, X_test_pca_minmax),
    'robust': (X_train_pca_robust, X_test_pca_robust)
}

results = []

for model_name, model_info in best_models.items():
    if isinstance(model_info, dict) and 'model' in model_info and 'version' in model_info:
        model = model_info['model']
        best_version = model_info['version']
        
        if best_version in pca_data_mapping:
            X_train_data, X_test_data = pca_data_mapping[best_version]
            
            y_train_pred = model.predict(X_train_data)
            y_test_pred = model.predict(X_test_data)
            
            train_metrics = {
                'Model': model_name,
                'Dataset': 'Train',
                'PCA_Version': best_version,
                'R2': r2_score(y_train, y_train_pred),
                'MAE': mean_absolute_error(y_train, y_train_pred),
                'RMSE': np.sqrt(mean_squared_error(y_train, y_train_pred)),
                'MAPE': mean_absolute_percentage_error(y_train, y_train_pred)
            }
            
            test_metrics = {
                'Model': model_name,
                'Dataset': 'Test',
                'PCA_Version': best_version,
                'R2': r2_score(y_test, y_test_pred),
                'MAE': mean_absolute_error(y_test, y_test_pred),
                'RMSE': np.sqrt(mean_squared_error(y_test, y_test_pred)),
                'MAPE': mean_absolute_percentage_error(y_test, y_test_pred)
            }
            
            results.append(train_metrics)
            results.append(test_metrics)

results_df = pd.DataFrame(results)

print("Таблица метрик лучших версий PCA по моделям:")
print(f"{'Модель':<20} {'Версия PCA':<12} {'Выборка':<8} {'R²':<8} {'RMSE':<12} {'MAE':<12} {'MAPE':<12}")

if not results_df.empty:
    results_sorted = results_df.sort_values(['Model', 'Dataset'])
    
    for idx, row in results_sorted.iterrows():
        print(f"{row['Model']:<20} {row['PCA_Version']:<12} {row['Dataset']:<8} "
              f"{row['R2']:7.4f}  {row['RMSE']:11.4f}  {row['MAE']:11.4f}  {row['MAPE']:11.4f}")
    
    test_results = results_df[results_df['Dataset'] == 'Test']
    
    overall_best = test_results.loc[test_results['RMSE'].idxmin()]
        
    print("\nЛучшая модель на тестовой выборке:")
    print(f"Модель: {overall_best['Model']}")
    print(f"Версия PCA: {overall_best['PCA_Version']}")
    print(f"R²: {overall_best['R2']:.4f}")
    print(f"RMSE: {overall_best['RMSE']:.4f}")
    print(f"MAE: {overall_best['MAE']:.4f}")
    print(f"MAPE: {overall_best['MAPE']:.4f}")
```

Это привело к тому, что оставшиеся признаки имеют сложные, нелинейные взаимодействия с целевой переменной "Гармония Бессмертия", которые линейные модели не способны уловить.
